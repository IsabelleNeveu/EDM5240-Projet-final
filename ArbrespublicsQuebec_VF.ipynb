Création d'une variable «Frêne» (sous-section)

```python
a["Espece"] = a["Espece"].str.replace("frêne","frene")
```

```python
frene = a.Espece.str.contains("frene")
```

Combien y a-t-il de frênes à Québec ?

```python
a[frene].Espece.value_counts()
```

```python
a[frene].Espece.value_counts().sum()
```

Pourcentage de frênes parmi l'ensemble des arbres publics à Québec : 

```python
a[frene].Espece.value_counts().sum()/len(a.Espece)*100
```
