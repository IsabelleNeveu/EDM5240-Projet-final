PROBLÈME: Lorsque je demande à Pandas d'afficher uniquement les espèces qui contiennent le nom «Frêne» et de dénombrer chacune de ces espèces, Pandas affiche l'erreur suivante : ValueError: cannot index with vector containing NA / NaN values (voir plus bas). 

J'ai essayé diverses tentatives pour tenter de résoudre le problème, mais toujours sans succès. 

J'ai donc opté pour faire cette opération dans EXCEL. 

1) J'ai utilisé la fonction NB.SI, afin de dénombrer le nombre de frênes dans la colonne «Espece_nom_français». 

2) Voici la formule utilisé :       =NB.SI(F:F;"*Frêne*")
J'ai entouré le mot Frêne de deux *, afin que la recherche prenne en considération l'ensemble des possibilités.

3) Résultat : un total de 9 326 frênes
